In this assignment we will be looking at how minute errors caused by the method in which a measurement was taken can result in non-trivial deviations to the derivative of the measurement vector.

Firstly, we will take a simple quadratic function, and plot another function in the same graph but with a small peaks. These peaks will be revealed by the function but should be almost unnoticable by just looking at the graph

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(0,3,0.001)

y1 = x**2 + 2
y2 = x**2 + 2 + (np.sin(200*x)/1000)

plt.plot(y1)
plt.plot(y2)
```

As we see, the functions, to the bare eye look like they overlap, but when we take the forward differentiation of these functions, similar to how some measurements can be differentiated in real life, we see something drastic.

```python
y1f = np.diff(y1)
y2f = np.diff(y2)

plt.plot(y1f)
plt.plot(y2f)
```
