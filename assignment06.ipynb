In this assignment we look at continuity, differentiation, taylor series and optimization.


- The basis of calculus lie in the functionalities of differentiation and integration. A function is continuous if the function does not have any abrupt changes. For example, if the function's slope tends to infinity or in a piecewise function the space between two continuing points has a slope of infinity. On the other hand, if the limit of a function does not exist, then the function is not continuous.

- And finally, if a function is differentiable throughout its domain, then it is also continuous at every point in its domain. But the opposite is not true.

- Now, lets take a look at derivatives. For an example, lets compute the derivative of `x^3+2x^2`. We know that the derivative of it is `3x^2+4x`.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3,1,0.012)

y = x**3 + 2*x**2

dy_forward = np.diff(y)/0.01

xd = x[:-1:]
dy_computed = 3*xd**2 + 4*xd
plt.plot(dy_computed)
plt.plot(dy_forward)


print(max(abs(dy_forward - dy_computed)))
```

- The graph taken a look at this way almost seems to have the two functions overlapping. As we see, there is an error as computed by the absolute of differences. This error would only increase as the interval between the two points in the declaration of x's range is increased.
- The forward derivative produces an error because the slope is slightly different as it is the slope between the point at x and the point x + 1 and not the slope at x. Therefore, if the difference in the slope at x and x+1 is more different, then the forward difference produces larger errors.
